```
import json
from google.colab import drive
drive.mount('/content/drive')
data = []
counter = 0

with open('/content/drive/My Drive/NLP/tifu_all_tokenized_and_filtered.json') as f:
  for line in f:
      data.append(json.loads(line))
      counter += 1
      if counter == 100:
        break

```

```
!pip install transformers
import torch
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
from nltk.tag import pos_tag
from nltk.tokenize import word_tokenize
from collections import Counter
from transformers import PegasusTokenizer, PegasusForConditionalGeneration
model_name = 'google/pegasus-reddit_tifu'
torch_device = 'cuda' if torch.cuda.is_available() else 'cpu'
tokenizer = PegasusTokenizer.from_pretrained(model_name)
model = PegasusForConditionalGeneration.from_pretrained(model_name).to(torch_device)
with open('data.txt', 'w') as outfile:
  for i in data:
    d = {}
    d['original_text'] = i['selftext']
    count= Counter([j for i,j in pos_tag(word_tokenize(i['selftext']))])
    src_text = []
    src_text.append(i['selftext'])
    batch = tokenizer.prepare_seq2seq_batch(src_text, truncation=True, padding='longest').to(torch_device)
    translated = model.generate(**batch)
    tgt_text = tokenizer.batch_decode(translated, skip_special_tokens=True)
    c= Counter([j for i,j in pos_tag(word_tokenize(tgt_text[0]))])
    d['summary_text'] = tgt_text[0]
    d['compression_ratio'] = len(tgt_text[0])/len(src_text[0])
    d['original_pos_tags'] = count
    d['summary_pos_tags'] = c
    json.dump(d, outfile)
    outfile.write('\n')
    print('JSON LINE ADDED')
      
```
